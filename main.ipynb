```python
import openpyxl
import re

# Create a new Excel workbook and select the active sheet
workbook = openpyxl.Workbook()
sheet = workbook.active

# Set the column headers
sheet.append([ 'Product Name', 'Price', 'Description','Image Link'])

# Assuming `product` is a list of parsed HTML elements (e.g., using BeautifulSoup)

for i in range(len(product)):
    # Extract image link
    img_tag = product[i].find('img')
    image_link = img_tag['src'] if img_tag and 'src' in img_tag.attrs else None

    # Extract product name
    product_name = img_tag['alt'] if img_tag and 'alt' in img_tag.attrs else None

    # Extract price
    price = product[i].find('div', {'class': "Nx9bqj _4b5DiR"})
    price_text = price.text if price else None

    # Extract and format description
    name = product[i].find('div', {'class': "_6NESgJ"})
    if name:
        desc = name.text
        # Format the description
        formatted_desc = desc.replace('|', ',')
        formatted_desc = re.sub(r'(\D)(\d)', r'\1 \2', formatted_desc)
        formatted_desc = re.sub(r'(\d)(\D)', r'\1 \2', formatted_desc)
        formatted_desc = re.sub(r'\s{2,}', ' ', formatted_desc).strip()
    else:
        formatted_desc = None

    # Append the data to the sheet
    sheet.append([product_name, price_text, formatted_desc,image_link])

# Save the workbook
excel_filename = 'product_details_direct.xlsx'
workbook.save(excel_filename)

print(f"Data has been successfully saved to {excel_filename}")
```

```python
import pandas as pd
import re

# Assuming `product` is a list of parsed HTML elements (e.g., using BeautifulSoup)

# Lists to store data
image_links = []
product_names = []
prices = []
descriptions = []

# Point 1: Extract image links
for i in range(len(product)):
    img_tag = product[i].find('img')
    if img_tag and 'src' in img_tag.attrs:
        image_links.append(img_tag['src'])
    else:
        image_links.append(None)  # To maintain alignment

# Point 2: Extract product names
for i in range(len(product)):
    img_tag = product[i].find('img')
    if img_tag and 'alt' in img_tag.attrs:
        product_names.append(img_tag['alt'])
    else:
        product_names.append(None)

# Point 3: Extract prices
for i in range(len(product)):
    price = product[i].find('div', {'class': "Nx9bqj _4b5DiR"})
    if price is not None:
        prices.append(price.text)
    else:
        prices.append(None)

# Point 4: Extract descriptions
for i in range(len(product)):
    name = product[i].find('div', {'class': "_6NESgJ"})
    if name is not None:
        desc = name.text
        # Format the description
        formatted_output = desc.replace('|', ',')
        formatted_output = re.sub(r'(\D)(\d)', r'\1 \2', formatted_output)
        formatted_output = re.sub(r'(\d)(\D)', r'\1 \2', formatted_output)
        formatted_output = re.sub(r'\s{2,}', ' ', formatted_output).strip()
        descriptions.append(formatted_output)
    else:
        descriptions.append(None)

# Create a DataFrame
data = {
    'Image Link': image_links,
    'Product Name': product_names,
    'Price': prices,
    'Description': descriptions
}
df = pd.DataFrame(data)

# Save to Excel
excel_filename = 'product.xlsx'
df.to_excel(excel_filename, index=False)

print(f"Data has been successfully saved to {excel_filename}")
```
